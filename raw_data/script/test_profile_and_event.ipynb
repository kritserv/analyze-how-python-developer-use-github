```python
import time
```

```python
import requests

session = requests.Session()
with open ("token.txt", "r") as f:
    x = f.readline()
session.headers.update({"Authorization": "token "+x.strip()})
```

```python
username = "kritserv"
```

```python
profile_url = "https://api.github.com/users/"+username
```

```python
response = session.get(profile_url)
if response.status_code == 200:
    profile_data = response.json()
    if profile_data["type"] == "Organization":
        print("Organization")
    else:
        print(profile_data["type"])
```

```python
event_data = {}

for i in [1,2,3]:
    url_pagination = "https://api.github.com/users/"+username+"/events?per_page=100&page="+str(i)
    response = session.get(url_pagination)
    if response.status_code == 200:
        event_data["page"+str(i)] = response.json()
        poll_interval = response.headers.get('X-Poll-Interval')
        print("wait", int(poll_interval)+5, "sec ...")
        time.sleep(int(poll_interval)+5)
    else:
        print("error", response.status_code)
        break
```
